##### Example 1-11. Summation of elements in Python

```python
x  = [1, 4, 6, 2]
n = len(x)

summation = sum(10*x[i] for i in range(0,n))
print(summation)
```

## Euler's Number and Natural Logarithms
##### Example 1-13. Calculating Compund Interest in Python

```python
from math import exp

p = 100
r = 0.20
t = 2.0
n = 12

a = p * (1 + (r/n))**(n * t)
print(a)
```

##### Example 1-14. Calculating continuous interest in Python

```python
p = 100 # Principal amount
r = 0.20 # Rate
t = 2.0 # Time

a = p * exp(r*t)
print(a)
```
